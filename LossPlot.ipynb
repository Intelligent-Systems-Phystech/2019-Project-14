```python
import re
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
logfile = 'reslog.txt'
```

```python
with open(logfile, 'r') as log:
    losses = log.readlines()
```

```python
losses = [re.split(' ', loss.replace('\n', '')) for loss in losses]
losses = [[float(loss[i]) for i in [2, 5]] for loss in losses]
losses = pd.DataFrame(losses)
losses = losses.rename(columns={0: 'epoch', 1: 'loss'})
losses = losses.groupby(by='epoch')['loss'].mean()
```

```python
plt.figure(figsize=(16,9))
plt.plot(losses, linestyle='--', marker='o')
```

```python
plt.figure(figsize=(16,9))
plt.plot(losses[20:50], linestyle='--', marker='o')
```

```python
plt.figure(figsize=(16,9))
plt.plot(losses[66:], linestyle='--', marker='o')
```
